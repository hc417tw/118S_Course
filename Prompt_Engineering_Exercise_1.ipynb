```python
# The AI I used is Google Gemini
# The prompt I request to the Gemini is "I need you to build an customer reflection AI chatbot, asking customer how they like about today's meals.
# A simple multi-step flow where the output of one prompt becomes input to the next. I need this in Python code so I can excute in colab and test it out by chating with it."
# In addition, I have request Gemini to add more steps for asking on customer serivce, so Gemini add steps on asking team members.

def extended_service_bot():
    print("--- ðŸŒŸ Premium Guest Relations Assistant ---")

    # STEP 1: Initial Sentiment
    mood = input("\nBot: Welcome! How was your dining experience today? (Excellent / Average / Disappointing): ").strip().lower()

    # STEP 2: The Deep Dive (Branching Logic)
    if "ex" in mood: # Excellent
        feedback = input("Bot: Thatâ€™s what we love to hear! What specifically made it stand out? ")
        service_context = "positive"
    elif "dis" in mood or "bad" in mood: # Disappointing
        feedback = input("Bot: Iâ€™m truly sorry we let you down. Was the issue with the food quality or the service speed? ")
        service_context = "negative"
    else:
        feedback = input("Bot: Thanks for the feedback. What is one small thing we could do to earn a 5-star rating from you next time? ")
        service_context = "neutral"

    # STEP 3: Service Recovery / Reward (Chaining the 'service_context')
    print(f"\n[System: Processing {service_context} resolution...]")
    if service_context == "negative":
        print(f"Bot: I understand that {feedback} ruined your experience. I'd like to make it right.")
        input("Bot: Can I offer you a digital voucher for a free appetizer on your next visit? (Yes/No): ")
    elif service_context == "positive":
        print(f"Bot: We are so glad you enjoyed the {feedback}!")
        input("Bot: Since you had a great time, would you be open to leaving a quick Google review? (Yes/No): ")
    else:
        print("Bot: Thank you for that suggestion. We take every note seriously.")

    # STEP 4: Staff Recognition
    staff_member = input("\nBot: Did any particular team member make your visit better today? (Name or 'No'): ")

    # STEP 5: Closing & Case ID
    import random
    case_id = random.randint(1000, 9999)

    print(f"\n--- Final Summary ---")
    if staff_member.lower() != "no":
        print(f"Action: We will ensure {staff_member} is recognized for their hard work!")

    print(f"Action: Your feedback regarding '{feedback}' has been logged.")
    print(f"Bot: Thank you! Your reference number is #{case_id}. Have a wonderful day!")

# Run the extended bot
extended_service_bot()
```
